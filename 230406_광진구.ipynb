### 클러스터링(이게 맞나...?)

```python
fig, ax = plt.subplots(figsize=(5,5))
광진구초등학교위도경도.plot.scatter(x='Longitude',y='Latitude',ax=ax)
for ind in 광진구초등학교위도경도.index:
    ax.annotate(광진구초등학교위도경도.loc[ind]['학교명'], (광진구초등학교위도경도.loc[ind]['Longitude'], 광진구초등학교위도경도.loc[ind]['Latitude']))
```

```python
광진구초등학교위도경도.index
```

```python
# clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,random_state=50).fit(광진구초등학교위도경도[['Longitude','Latitude']])
labels=kmeans.labels_
centers=kmeans.cluster_centers_
```

```python
광진구초등학교위도경도['clustering']=labels
광진구초등학교위도경도.head(15)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

df_temp=광진구초등학교위도경도[광진구초등학교위도경도['clustering']==0]
df_temp.plot.scatter(x='Longitude',y='Latitude',ax=ax,color='RED')

df_temp=광진구초등학교위도경도[광진구초등학교위도경도['clustering']==1]
df_temp.plot.scatter(x='Longitude',y='Latitude',ax=ax,color='Orange')

df_temp=광진구초등학교위도경도[광진구초등학교위도경도['clustering']==2]
df_temp.plot.scatter(x='Longitude',y='Latitude',ax=ax,color='Green')

df_temp=광진구초등학교위도경도[광진구초등학교위도경도['clustering']==3]
df_temp.plot.scatter(x='Longitude',y='Latitude',ax=ax,color='Blue')

for ind in 광진구초등학교위도경도.index:
    ax.annotate(광진구초등학교위도경도.loc[ind]['학교명'], (광진구초등학교위도경도.loc[ind]['Longitude'], 광진구초등학교위도경도.loc[ind]['Latitude']))
```
